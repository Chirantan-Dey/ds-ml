```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("ds.csv")
df.head()
```

```python
countries = ['IN', 'ID', 'MY', 'VN', 'TH', 'PH', 'PL', 'UA', 'RO', 'HU', 'BR', 'MX', 'CO', 'AR', 'ZA', 'KE', 'NG', 'EG']
df = df[(df['employment_type'] == 'FT') & (df['employee_residence'].isin(countries))]
df.head()
```

```python
df = df.drop(columns=['employment_type',
                    'employee_residence',
                    'salary','salary_currency',
                    'company_location',
                    'job_title',
                    'remote_ratio'])
df.head()
```

```python
df.loc[df['work_year'].isin([2020, 2021]), 'salary_in_usd'] *= 1.25
df['salary_in_usd'] = df['salary_in_usd'].round(2)
df.head()
```

```python
df = df.drop(columns=['work_year'])
df.head()
```
